```python
def my_f_1(a,b):
    assert a>0;" intensity pozitive","error intensity not pozitive"
    if(a<=255-b):
        return a+b
    else:
        return 255
my_f_1(243,30)    
```

```python
def my_f_2(a):
    #assert a>=0;"intensity positive","error intensity not positive"
    return int(255-a)
m_f_2(234)
```

```python
def get_distance(v,w=[1/3,1/3,1/3]):
    a,b,c=v[0],v[1],v[2]
    w1,w2,w3=w[0],w[1],w[2]
    d=((a**2)*w1+ 
      (b**2)*w2+
      (c**2)*w3)**.5
    return d
    


def convert_rgb_to_gray_level(im_1):
    m,n,k = im_1.shape
    new1 = np.zeros((m,n),dtype='uint8')
    for i in range(m):
        for j in range(n):
            new_image=get_distance(im_1[i,j,:])
            new1[i,j]=new_image
    return new1


```
