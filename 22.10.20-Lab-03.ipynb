```python
def get_value_from_triple(temp_1):
    #temp_1 = im_1[0,0,:]
    return int(temp_1[0]/3 + temp_1[1]/3+temp_1[2]/3)


def get_0_1_value_from_triple(temp_1):
    #temp_1 = im_1[0,0,:]
    temp=int(temp_1[0]/3 + temp_1[1]/3+temp_1[2]/3)
    if temp<110:
        return 0
    else:
        return 1
    
    
def convert_rgb_to_bw(im_1):
    m,n,k = im_1.shape
    new_image = np.zeros((m,n),dtype='uint8')
    for i in range (m):
        for j in range(n):
            s=get_0_1_value_from_triple(im_1[i,j,:])
            new_image[i,j]=s
    return new_image  

def convert_rgb_to_gray(im_1):
    m,n,k = im_1.shape
    new_image = np.zeros((m,n),dtype='uint8')
    for i in range(m):
        for j in range(n):
            s = get_value_from_triple(im_1[i,j,:])
            new_image[i,j]=s
    return new_image   
```

```python
get_value_from_triple(im_1[10,10,:])
get_0_1_value_from_triple(im_1[10,10,:])
im_3 = convert_rgb_to_bw(im_1)
im_2 = convert_rgb_to_gray(im_1)
```

```python
plt.subplot(1,3,1)
plt.imshow(im_1)
plt.subplot(1,3,2)
plt.imshow(im_2,cmap='gray')
plt.subplot(1,3,3)
plt.imshow(im_3,cmap='gray')
plt.show()
```

```python
plt.imsave('canakkale_gray.jpg',im_2,cmap='gray')
plt.imsave('canakkale_bw.jpg',im_3,cmap='gray')
```
