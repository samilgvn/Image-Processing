```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compare_list_ndarray():
    l_1 = [1,"aaaaaa,3",'4',5,6] #broadcast
    l_2 = [2,"2adadadada,3",'1114',15,26] # broadcast
    print(l_1+l_2) #+

    list1 = [1,2,3,4]
    list2 = [1,2,3,4]
    print(list1+list2 + [10]) # + list_1+list_2+10 error

    list3 = np.asarray([1,2,3,4])
    list4 = np.asarray([1,2,3,4])
    print(list3+list4+10)

def get_jpg_files():
    os.getcwd()
    os.listdir()
    path = os.getcwd()
    jpg_files =[f for f in os.listdir(path) if f.endswith('.jpg')]
    return jpg_files


compare_list_ndarray()
get_jpg_files()
```

```python
def my_display_two_image(im1,im2):
    plt.subplot(1,2,1)
    plt.imshow(im1)
    plt.subplot(1,2,2)
    plt.imshow(im2)

    plt.show()

def my_rotate(im_1): 
    m,n,k =im_1.shape
    new_image = np.zeros((840,471,3),dtype="uint8")
    for i in range(m):
        for j in range(n):
            temp1 = im_1[i,j]
            new_image[j,i]=temp1
    return new_image
```

```python
im_1 = plt.imread('pic_1.jpg')
```

```python
im_2 = my_rotate(im_1)
my_display_two_image(im_1,im_2)
```
